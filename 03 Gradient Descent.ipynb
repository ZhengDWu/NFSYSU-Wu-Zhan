### 实现梯度下降

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 使用 numpy 构造 虚拟 数据

```python
x = np.linspace(-10, 5, 151)
y = (x+2.5)**2 + 7# loss function
```

```python
plt.plot(x, y)
plt.show()
```

### 实现梯度下降

```python
# 求导
def dJ(theta):
    return 2*(theta+2.5)
```

```python
# loss function
def J(theta):
    return (theta+2.5)**2 + 7
```

```python
def gradient_descent(lr, theta):
    ''' 传入 学习率lr， theta 的初始位置'''
    limit = 1e-7 # 由于不太可能让导数跑到 0，所以加个判定条件，证明损失函数已经收敛了
    theta_histories = [theta] # 记录 theta 的走向
    while True:
        direction = dJ(theta) # 求导找方向
        last_theta = theta # 保存一开是 theta 值
        theta = theta - lr * direction # 向 “谷底” 移动一步，生成新的 theta 值
        theta_histories.append(theta) # 持续添加 每次移动后的 thata
        
        if(abs(J(theta) - J(last_theta)) < limit):
            break
    return theta, theta_histories
```

```python
lr = 0.1 # 学习率
theta= -8.0

theta, theta_histories = gradient_descent(lr, theta)
print(theta)
print(J(theta))
```

```python
plt.plot(x, y)
plt.plot(np.array(theta_histories), J(np.array(theta_histories)), 'r+')
```

```python
print('learning rate ：', lr, '移动次数: ', len(theta_histories))
```

### 改变 learning rate 试试

```python
lr = 0.01 # 学习率
theta= -8.0

theta, theta_histories = gradient_descent(lr, theta)
print(theta)
print(J(theta))
plt.plot(x, y)
plt.plot(np.array(theta_histories), J(np.array(theta_histories)), 'r+')
```

```python
print('learning rate ：', lr, '移动次数: ', len(theta_histories))
```
