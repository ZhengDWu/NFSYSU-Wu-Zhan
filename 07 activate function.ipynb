## 激活函数实现

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 创建虚拟数据
x = np.linspace(-7, 7, 150)
```

### 动手实现激活函数

```python
def sigmoid(xs):
    y = [1 / float(1 + np.exp(-x)) for x in xs]
    return y
```

```python
def relu(xs):
    y = [x * (x > 0) for x in xs]
    return y
```

```python
def tanh(xs):
    y = [(np.exp(x) - np.exp(-x)) / float(np.exp(x) + np.exp(-x))
        for x in xs]
    return y
```

```python
def softplus(xs):
    y = [np.log(1 + np.exp(x)) for x in xs]
    return y
```

```python
y_sigmoid = sigmoid(x)
y_relu = sigmoid(x)
y_tanh = tanh(x)
y_softmax = softplus(x)

plt.figure(1, figsize = (8, 6))
plt.subplot(221)
plt.plot(x, y_sigmoid, c = 'red', label = 'sigmoid')
plt.legend(loc = 'best')

plt.figure(1, figsize = (8, 6))
plt.subplot(222)
plt.plot(x, y_sigmoid, c = 'red', label = 'relu')
plt.legend(loc = 'best')

plt.figure(1, figsize = (8, 6))
plt.subplot(223)
plt.plot(x, y_tanh, c = 'red', label = 'tanh')
plt.legend(loc = 'best')

plt.figure(1, figsize = (8, 6))
plt.subplot(224)
plt.plot(x, y_softmax, c = 'red', label = 'softmax')
plt.legend(loc = 'best')
```

### 使用 torch 自带的激活函数 （需要将 numpy 转为 tensor 才能使用）

```python
import torch
import torch.nn.functional as F
from torch.autograd import Variable

x = torch.linspace(-5, 5, 200)
x = Variable(x)
np_x = x.data.numpy()

# 调用 torch 自带的 激活函数
y_relu = F.relu(x).data.numpy()
y_sigmoid = F.sigmoid(x).data.numpy()
y_tanh = F.tanh(x).data.numpy()
y_softmax = F.softplus(x).data.numpy()
```

```python
plt.figure(1, figsize = (8, 6))
plt.subplot(221)
plt.plot(np_x, y_sigmoid, c = 'red', label = 'sigmoid')
plt.legend(loc = 'best')

plt.figure(1, figsize = (8, 6))
plt.subplot(222)
plt.plot(np_x, y_sigmoid, c = 'red', label = 'relu')
plt.legend(loc = 'best')

plt.figure(1, figsize = (8, 6))
plt.subplot(223)
plt.plot(np_x, y_tanh, c = 'red', label = 'tanh')
plt.legend(loc = 'best')

plt.figure(1, figsize = (8, 6))
plt.subplot(224)
plt.plot(np_x, y_softmax, c = 'red', label = 'softmax')
plt.legend(loc = 'best')
```
